# JD SCRAPPING

```python
from time import sleep
import time

from selenium import webdriver
import pandas as pd 
from bs4 import BeautifulSoup
options = webdriver.ChromeOptions()
from webdriver_manager.chrome import ChromeDriverManager
driver = webdriver.Chrome(ChromeDriverManager().install())
df = pd.DataFrame(columns=["Title","Location","Experience","Company","Description"])
for i in range(1,2000,1):
    driver.get('https://www.shine.com/job-search/business-analytics-jobs-'+str(i))
    jobs = []
    driver.implicitly_wait(4)

    for job in driver.find_elements_by_class_name('result-display__profile'):
        soup = BeautifulSoup(job.get_attribute('innerHTML'),'html.parser')

        try:
            title = soup.find("a",class_="job_title_anchor").text.replace("\n","").replace("\xa0"," ").strip()
        except:
            title = 'None'
        try:
            location = soup.find("ul",class_="d-flex flex-row mt-10").text.replace("\n","").strip()
        except:
            location = 'None'

        try:
            company = soup.find("span",class_="result-display__profile__company-name").text.replace("\n","").strip()
        except:
            company = 'None'

        sum_div = job.find_elements_by_class_name("result-display__profile__job-tittle")[0]
        try:
            sum_div.click()
        except:
            break
            
        time.sleep(1)
        job_desc = driver.find_element_by_id('jd_snippet_ajax').text

        df = df.append({'Title':title,'Location':location[50:].strip(), 'Experience':location[:20].strip(),"Company":company, "Description":job_desc},ignore_index=True)
        
```

```python
nan_value = float("NaN")
df.replace("", nan_value, inplace=True)

df = df.drop_duplicates()
df.dropna(subset = ["Description"], inplace=True)
df = df.reset_index()
df = df.drop(['index'], axis=1)
df.to_csv('shine_bi_jd0.csv')
df
```
