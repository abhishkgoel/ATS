# Load and Refine scraped data 

```python
import pandas as pd
df = pd.read_csv("shine_bi_jd0.csv")
```

```python
#df = df.reset_index()
df = df.drop(['Unnamed: 0'], axis=1)
df = df.drop_duplicates()
```

```python
#df.head(100)
```

```python
import re
exp = []
for i in range(len(df)):
    array = re.findall(r"\d{1,2}", df['Experience'][i])
    if(len(array)!=0):
        exp.append(array)
        
    else:
        exp.append(0)
        
#exp
```

```python
#len(exp)
```

# DESIGNATION

```python
import re
desg=[]
for i in range(len(df)):
    pair=[]
    asd = exp[i]
    for words in ['1','2']:
        if((asd)==0):
            pair.append('Fresher')
            break
        elif re.search(r'\b' + words + r'\b', str(asd)):
            pair.append('Entry level')        
    for words in ['3','4']:
        if re.search(r'\b' + words + r'\b', str(asd)):
            pair.append('Junior level')
    for words in ['5','6','7']:
        if re.search(r'\b' + words + r'\b', str(asd)):
            pair.append('Mid Senior level')
    for words in ['8','9','10','11','12']:
        if re.search(r'\b' + words + r'\b', str(asd)):
            pair.append('Senior level')
    for words in ['13','14','15','16']:
        if re.search(r'\b' + words + r'\b', str(asd)):
            pair.append('Director level')
    for words in ['17','18','19']:
        if re.search(r'\b' + words + r'\b', str(asd)):
            pair.append('Executive level')
    if(len(pair)==1):
        desg.append(pair)
    elif(len(pair)==2):
        desg.append([pair[1]])
    else:
        desg.append('C Level')
#(desg)
```

```python
#len(desg)
```

```python
df['Designation']=desg
```
