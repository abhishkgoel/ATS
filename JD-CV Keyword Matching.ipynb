# JD-CV Keyword Matching

```python
import pandas as pd
df = pd.read_csv(r"C:\Users\Abhishek_2\jupiter notebook\shineds0.csv")
df = df.drop('Unnamed: 0', axis = 1)
Resume_data = pd.read_csv(r"C:\Users\Abhishek_2\Desktop\cv2\resume0.csv")
Resume_data = Resume_data.drop('Unnamed: 0', axis = 1)
```

```python
import re
import ast

skills1 = ast.literal_eval(df['TechSkillMatch'][1]) #for converting string list to list
skills1 = list(skills1.keys())
resumemutli=[]
for j in range(5):
    tt0 = Resume_data['cv_text'][j]
    tt1 = tt0.replace('\n','')
    tt1 = tt1.lower()
    resume = []
    for i in range(len(skills1)):
        asd = re.findall(r'\b' + skills1[i] + r'\b',tt1)
        resume.append(asd)
    

    #resume = pd.DataFrame(resume)
    resumemutli.append(resume)
#resumemutli

Resumemutli1 = []
for i in range(len(resumemutli)):
    Resumemutli = []
    for j in range(len(resumemutli[i])):
        if(len(resumemutli[i][j])==0):
            Resumemutli.append(0)
        else:
            Resumemutli.append(len(resumemutli[i][j]))
    Resumemutli1.append(Resumemutli)
#Resumemutli1

import ast
skills1 = ast.literal_eval(df['TechSkillMatch'][1]) #for converting string dict to dict
new = pd.DataFrame.from_dict([skills1])
new = new.transpose()
new.columns = ['job_desc']
#new

for i in range(len(Resumemutli1)):
    new['Resume',i]= Resumemutli1[i]
new
```
